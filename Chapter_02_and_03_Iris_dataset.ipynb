## Chapter 3 – Cleaning and Describing Data Using Python

```
# Load the libraries needed to perform Exercise 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the Iris dataset with missing values

Replace the file path below with the location of your `Iris_m.csv` file.


```
# Load the data
data = pd.read_csv(r'C:\Path of your file\Iris_m.csv')

# Display first five rows
data.head()
```

```
# len() function can tell how many samples are in the dataset (rows)
len(data)
```

```
# isna() function can identify empty cells in the data
data.isna()
```

```
# Adding .sum() after isna() will output the number of empty cells
data.isna().sum()
```

```
# Coercing the data to numeric will also convert all invalid data to empty cells
data['Sepal_width'] = data['Sepal_width'].apply(
    pd.to_numeric,
    errors='coerce'
)

# Check missing values again
data.isna().sum()
```

### Cleaning Missing Values

```
# Define the variables to consider in EDA
variables = [
    'Petal_width',
    'Petal_length',
    'Sepal_width',
    'Sepal_length'
]

# Drop the missing or invalid data
cleandf = data.dropna(subset=variables)

# Create a new object to store the cleaned data
df = cleandf

# Display first five rows
df.head()
```

```
# Check dataset shape after cleaning
df.shape
```

### Correcting Invalid Species Names

```
# Correct the invalid species names
df = df.replace('Verginica', 'Virginica', regex=True)

# Remove the Species_No column
df = df.drop('Species_No', axis=1)

# Display first five rows
df.head()
```

### Descriptive Statistics

```
# Perform descriptive statistics on the Iris dataset
df.describe()
```

```
# Perform descriptive statistics for each Iris species
dtable = df.groupby(['Species_name']).describe()

# Display descriptive statistics
dtable
```

```
# Transpose the descriptive statistics table
dtable = dtable.transpose()

# Display transposed table
dtable
```

### Histogram Plots

```
# Plot histogram for Petal Length
plt.figure(figsize=(8, 5))
plt.hist(df['Petal_length'], bins=10)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()
```

```
# Plot histogram for Sepal Length
plt.figure(figsize=(8, 5))
plt.hist(df['Sepal_length'], bins=10)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

### Pairplot Visualization

```
# Pairplot visualization of the Iris dataset
sns.pairplot(df, hue='Species_name')
plt.show()
```
